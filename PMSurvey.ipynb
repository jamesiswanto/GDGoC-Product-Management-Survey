# **Survey of Product Management in Google Developer Groups on Campus - Universitas Gunadarma**

```python
# Set Up the Environment and Import Libraries
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display, HTML

# Set seaborn style for better aesthetics
sns.set(style="whitegrid")
```

```python
# Load the Dataset
# Load the CSV file
file_path = "https://raw.githubusercontent.com/jamesiswanto/GDGoC-Product-Management-Survey/refs/heads/development/PMSurvey.csv"
data = pd.read_csv(file_path)

# Display first few rows to understand structure
data.head()
```

```python
# Define total respondents
total_respondents = data.shape[0]

# Set up the figure and aesthetics
fig, ax = plt.subplots(figsize=(4, 2), dpi=80)
fig.patch.set_facecolor('white')     # Outer background color for contrast

# Display text in the center of the plot
ax.text(0.5, 0.6, str(total_respondents),
        fontsize=36, fontweight='bold', color="black", ha='center', va='center')
ax.text(0.5, 0.2, "Total Respondents",
        fontsize=16, fontweight='bold', color="black", ha='center', va='center')

# Remove axes for a clean look
ax.axis('off')

# Show and save the figure as an image
plt.show()
```

```python
# Visualization of Majors
# Count the number of respondents for each major
plt.figure(figsize=(10, 6))
major_counts = data['Major \nJurusan '].value_counts()
sns.barplot(x=major_counts.values, y=major_counts.index, palette="viridis")

plt.title("Distribution of Respondents' Majors")
plt.xlabel("Number of Respondents")
plt.ylabel("Major")
plt.show()
```

```python
# Visualization of Experience Level
# Count the number of responses for experience level
plt.figure(figsize=(8, 5))
experience_counts = data['Do you have any prior experience or knowledge about Product Management?\nApakah Anda memiliki pengalaman atau pengetahuan sebelumnya tentang Product Management?'].value_counts()
sns.barplot(x=experience_counts.index, y=experience_counts.values, palette="plasma")

plt.title("Respondents' Prior Experience in Product Management")
plt.xlabel("Experience Level")
plt.ylabel("Number of Respondents")
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualization of Interest in Weekly Product Management Class
# Pie chart of interest level in the weekly PM class
plt.figure(figsize=(8, 8))
interest_counts = data['How interested are you to join Weekly Product Management Class?\nSeberapa tertarik Anda untuk mengikuti weekly class Manajemen Produk?'].value_counts()
plt.pie(interest_counts, labels=interest_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("coolwarm", len(interest_counts)))

plt.title("Interest Level in Weekly Product Management Class")
plt.show()
```
